```python
from sklearn.datasets import load_iris
from sklearn import tree

iris = load_iris()
```

```python
X, y = iris.data, iris.target
clf = tree.DecisionTreeClassifier()
clf.fit(X, y)
tree.plot_tree(clf)
```

```python
 import graphviz
 dot_data = tree.export_graphviz(clf, out_file = None)
 graph = graphviz.Source(dot_data)
 graph.render("iris")
```

```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

cancer_data = load_breast_cancer(as_frame = True) #loaded as dataframe
print(cancer_data.DESCR) #describes sataset
```

```python
 X, y = load_breast_cancer(return_X_y = True)
 X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0)
```

```python
from pandas.core.common import random_state
clfb = tree.DecisionTreeClassifier()
clfb.fit(X_train, y_train)
prediction = clfb.predict(X_test)

for xi, yi in zip(prediction, y_test):
  print(xi, yi)
```

```python
path = clfb.cost_complexity_pruning_path(X_train, y_train)
ccp_alpha, impurities = path.ccp_alphas, path.impurities
```
