```python
import pandas as pd

address = 'https://opendata.cern.ch/record/5202/files/Dimuon_SingleMu.csv'
df = pd.read_csv(address)
```

```python
list(df.columns)
```

```python
import matplotlib.pyplot as plt
plt.hist(df.M, bins=100, range=[0,150])
plt.show()
```

```python
import numpy as np

array = np.linspace(0,100,101)
plt.hist(df.M, bins=array, label='CMS data 2011')
plt.xlabel('Dimuons mass [GeV/$c^2$]')
plt.ylabel('Events')
plt.legend()
plt.show()
```

```python
selected_df = df.query('M < 5').copy()
plt.hist(df.M, bins=np.linspace(0,10,201), label='CMS data 2011')
plt.xlabel('Dimuons mass [GeV/$c^2$]')
plt.ylabel('Events')
plt.hist(selected_df.M, bins=np.linspace(0,5,101), label='Low mass')
plt.legend()
plt.show()
```

```python
selected_df.query("pt1 > 10 and pt2 > 10", inplace=True)
plt.hist(df.M, bins=np.linspace(0,10,201), label='CMS data 2011')
plt.xlabel('Dimuons mass [GeV/$c^2$]')
plt.ylabel('Events')
plt.hist(selected_df.M, bins=np.linspace(0,5,101), label='Selected data')
plt.legend()
plt.show()
```

```python
selection_condition = (df.pt1 > 10) & (df.pt2 > 10)
plt.hist(df.M[selection_condition==True],  bins=np.linspace(0,5,101), label='Selcted data')
plt.legend()
plt.show()
```
