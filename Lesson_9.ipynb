```python
import numpy as np
from sklearn.datasets import load_iris
from sklearn.linear_model import Perceptron

iris = load_iris()
x = iris.data
y = (iris.target == 0).astype(int)
```

```python
per_clf = Perceptron()
per_clf.fit(x, y)
```

```python
per_clf.score(x, y)
```

```python
print(per_clf.predict([[5.3, 3.4, 2.3, 0.4]]))
```

```python
import tensorflow as tf
from tensorflow import keras
print(tf.__version__)
print(keras.__version__)
```

```python
model = keras.models.Sequential(name='iris_1')
model.add(keras.layers.InputLayer(4))
model.add(keras.layers.Dense(10, activation = 'relu'))
model.add(keras.layers.Dense(4, activation = 'relu'))
model.add(keras.layers.Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
keras.utils.plot_model(model)
```

```python
model.compile(loss='binary_crossentropy', optimizer='sgd', metrics=['binary_accuracy'])
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(x, y)
```

```python
history = model.fit(X_train, y_train, epochs=30, validation_data=(X_test, y_test))
```

```python
import matplotlib.pyplot as plt

X_setosa = iris.data[iris.target==0]
plt.hist(model.predict(X_test), bins=50, range=(0,1))
plt.hist(model.predict(X_setosa), bins=50, range=(0,1), alpha=0.5)
plt.show()
```
